```python
import json
import matplotlib.pyplot as plt

# Load JSONL data
data = []
with open("../out/triples.jsonl", "r") as f:
    for line in f:
        data.append(json.loads(line))

# Count words in "positive"
word_counts_pos = []
word_counts_neg = []
for c in data:
    word_counts_pos.append(len(c["positive"].split()))
    word_counts_neg.append(len(c["negative"].split()))

# Plot histogram
plt.hist(word_counts_pos, bins=30, color='skyblue', edgecolor='black')
plt.title("Word Count in 'positive'")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot histogram
plt.hist(word_counts_neg, bins=30, color='skyblue', edgecolor='black')
plt.title("Word Count in 'negative'")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.show()
```
